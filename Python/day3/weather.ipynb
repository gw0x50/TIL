# 기상청 정보 읽기 (기상청 사이트 하단 RSS)

```python
import urllib.request as req
from bs4 import BeautifulSoup as bs
import ssl

url="https://www.weather.go.kr/weather/forecast/mid-term-rss3.jsp?stnId=108"
context=context = ssl._create_unverified_context()
response = req.urlopen(url,context=context)
soup = bs(response, "html.parser")

#print(soup.prettify())
'''
<location>
<city>서울
<wf>날씨
<tmx>최고기온
<tmn>최저기온
'''
city_list = soup.select("city")
wf = soup.select_one("wf")
print(len(city_list))
for city in city_list :
    print()
    wf = soup.select_one("wf")
    print(city.string + " : " + wf.string)
```

```python
import urllib.request as req
from bs4 import BeautifulSoup as bs
import ssl

url="https://www.weather.go.kr/weather/forecast/mid-term-rss3.jsp?stnId=108"
context=context = ssl._create_unverified_context()
response = req.urlopen(url,context=context)
soup = bs(response, "html.parser")

for loc in soup.select("location") :
    print("----------------------------------------")
    print("도시 : ", loc.select_one("city").string)
    print("시간 : ", loc.select_one("tmEf").string)
    print("날씨상황 : ", loc.select_one("wf").string)
    print("최고기온 : ", loc.select_one("tmx").string)
    print("최저기온 : ", loc.select_one("tmn").string)
    print("----------------------------------------")
```

```python
import urllib.request as req
from bs4 import BeautifulSoup as bs
import ssl
import matplotlib.pyplot as plt # 그래프
import matplotlib.font_manager as fm # 그래프 글꼴

url="https://www.weather.go.kr/weather/forecast/mid-term-rss3.jsp?stnId=108"
context=context = ssl._create_unverified_context()
response = req.urlopen(url,context=context)
soup = bs(response, "html.parser")

city_list = []
tmx_list = []
tmn_list = []

for loc in soup.select("location") :
    city_list.append(loc.select_one("city").string)
    tmx_list.append(loc.select_one("tmx").string)
    tmn_list.append(loc.select_one("tmn").string)

tmx_list = list(map(int, tmx_list)) # map으로 정렬
tmn_list = list(map(int, tmn_list)) # map으로 정렬

# 글꼴 세팅
plt.rcParams["font.family"] = "D2Coding"
plt.rcParams["font.size"] = 20
plt.rcParams["figure.figsize"] = (20,10)

plt.rcParams["xtick.labelsize"] = 12 # x축 데이터 글씨 크기
plt.rcParams["axes.labelsize"] = 16 # x축 테이블 글씨 크기
plt.rcParams["lines.linewidth"] = 3 # 그래프 선 두께
plt.rcParams["lines.linestyle"] = "-." # 그래프 선 모양
plt.rcParams["axes.grid"] = True # 눈금선 표시

# x축 도시명 y축 최고기온 그래프
plt.plot(city_list, tmx_list) # x축 y축
plt.title("도시별 기온") # 차트 위 제목
plt.xlabel("도시 이름") # x축 값 설명
plt.ylabel("기온") # y축 값 설명
plt.show()
```

```python
import urllib.request as req
from bs4 import BeautifulSoup as bs
import ssl
import matplotlib.pyplot as plt # 그래프
import matplotlib.font_manager as fm # 그래프 글꼴

url="https://www.weather.go.kr/weather/forecast/mid-term-rss3.jsp?stnId=108"
context=context = ssl._create_unverified_context()
response = req.urlopen(url,context=context)
soup = bs(response, "html.parser")

city_list = []
tmx_list = []
tmn_list = []

for loc in soup.select("location") :
    city_list.append(loc.select_one("city").string)
    tmx_list.append(loc.select_one("tmx").string)
    tmn_list.append(loc.select_one("tmn").string)

tmx_list = list(map(int, tmx_list)) # map으로 정렬
tmn_list = list(map(int, tmn_list)) # map으로 정렬

# 글꼴 세팅
plt.rcParams["font.family"] = "D2Coding"
plt.rcParams["font.size"] = 20
plt.rcParams["figure.figsize"] = (20,10)

plt.rcParams["xtick.labelsize"] = 12 # x축 데이터 글씨 크기
plt.rcParams["axes.labelsize"] = 16 # x축 테이블 글씨 크기
plt.rcParams["lines.linewidth"] = 3 # 그래프 선 두께
plt.rcParams["lines.linestyle"] = "-." # 그래프 선 모양
plt.rcParams["axes.grid"] = True # 눈금선 표시

# x축 도시명 y축 최고기온 그래프
plt.plot(city_list, tmx_list) # x축 y축
plt.title("도시별 기온") # 차트 위 제목
plt.xlabel("도시 이름") # x축 값 설명
plt.ylabel("기온") # y축 값 설명
plt.show()

# 1개 그래프(축 공유)에 도시별 최고기온과 최저기온 동시에 표현
plt.subplots()
plt.plot(city_list, tmx_list) # x축 y축
plt.plot(city_list, tmn_list) # x축 y축
plt.savefig("weather.png")
plt.show()
```
