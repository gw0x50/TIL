```python
# 파일을 한 라인씩 읽어서 리스트에 저장
file_list = list()
file3 = open("module.ipynb", "r", encoding="UTF-8")
for line in file3 :
    file_list.append(line)
file3.close()

# 리스트에 저장 내용 화면 출력
linenum = 1
for line in file_list :
    print(linenum, "번 라인 : ", line, end=" ")
    linenum += 1
```

```python
# 파일을 한 라인씩 읽어서 리스트에 저장
file_list = list()
file3 = open("module.ipynb", "r", encoding="UTF-8")
for line in file3 :
    file_list.append(line)
file3.close()

file4 = open("copy.txt", "w")

for index in range(0, len(file_list), 1) :
    line = str(index + 1) + " 번 라인 : " + file_list[index]
    file4.write(line)
file4.close()
```
