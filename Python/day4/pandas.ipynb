## 행, 열 합침

```python
df1 = pd.DataFrame ({
    '고객번호' : [1001, 1002, 1003, 1004,  1005, 1006, 1007],
    '이름' : ['둘리', '도우너', '또치', '길동', '희동', '마이콜', '영희']
})
df1
```

```python
df2 = pd.DataFrame ({
    '고객번호' : [1001, 1001, 1005, 1006, 1008, 1001],
    '금액' : [10000, 2000, 15000, 5000, 10000, 30000]
})
df2
```

```python
pd.merge(df1, df2) # 고객번호 존재하는 값을 가졌을 경우에만
```

```python
df1 # merge 결과는 원본에 반영되지 않음
```

```python
pd.merge(df1, df2, how = 'outer') # df1, df2 둘 다 표시
```

```python
pd.merge(df1, df2, how = 'left') # df1 기준 표시
```

```python
pd.merge(df1, df2, how = 'right') # df2 기준 표시
```

```python
df3 = pd.merge(df1, df2, how = 'outer')
df3.info()
```

```python
df3.isnull() # None 데이터 확인
```

```python
print(df3.isnull().sum()) # None 데이터 갯수 확인
```

```python
df4 = df3['금액'].fillna(0) # 금액의 None 값을 0으로 채움
df4
```

```python
df5 = df3['이름'].fillna("==") # 이름의 None 값을 ==으로 채움
df5
```

```python
df3.dropna() # None 데이터 있는 행 전부 삭제
```

```python
df3.dropna(axis=1) # None 데이터 있는 열 전부 삭제
```
