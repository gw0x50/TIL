### 인구 현황 데이터 파일 읽어오기

```python
pop_seoul = pd.read_excel('population_in_Seoul.xls', header = 2, usecols = ('B, D, G, J, N'))
pop_seoul.head()
```

#### 컬럼 구별, 인구수, 한국인, 외국인, 고령자 로 변경

```python
pop_seoul.rename(columns = { 
    pop_seoul.columns[0] : '구별',
    pop_seoul.columns[1] : '인구수',
    pop_seoul.columns[2] : '한국인',
    pop_seoul.columns[3] : '외국인',
    pop_seoul.columns[4] : '고령자'
}, inplace = True)

pop_seoul.head()
```

#### 구별 na값 삭제
##### 0번 인덱스 행의 데이터는 총 합계이므로 필요 없으니 삭제

```python
drop_seoul = pop_seoul.drop(0)
drop_seoul
```

##### '구별'의 데이터 값 확인

```python
drop_seoul['구별']
```

##### 데이터 확인 후 Nan 제거

```python
drop_seoul2 = drop_seoul.drop(26, axis = 0)
drop_seoul2
```

#### 외국인 비율 열을 추가 : 인구수 대비 외국인 비율 계산

```python
result_seoul = drop_seoul2
```

```python
result_seoul['외국인 비율'] = (result_seoul['외국인'] / result_seoul['인구수']) * 100
result_seoul
```

#### 고령자 비율 열을 추가 : 인구수 대비 고령자 비율을 계산

```python
result_seoul['고령자 비율'] = (result_seoul['고령자'] / result_seoul['인구수']) * 100
result_seoul
```

#### 인구수 가장 많은 상위구 5개를 조회

```python
result_seoul.sort_values('인구수', ascending = False).head() # head() 기본 값은 5
```

#### 외국인 가장 많은 상위구 5개를 조회

```python
result_seoul.sort_values('외국인', ascending = False).head()
```

#### 고령자 가장 많은 상위구 5개를 조회

```python
result_seoul.sort_values('고령자', ascending = False).head()
```

## 데이터 프레임 병합 및 전처리

### 데이터 병합

```python
data_result = pd.merge(CCTV_seoul, result_seoul)
data_result.head()
```

### 전처리
#### 인덱스 설정

```python
index_result = data_result.set_index('구별')
index_result
```

#### 불필요한 열 제거

```python
drop_columns = ['2013년도 이전', '2014년', '2015년', '2016년']
pop_result = index_result.drop(drop_columns, axis = 1)
pop_result
```

## 인구수와 소계 피벗 테이블

## 상관관계 분석 y = ax + b

## 그래프

### 그래프 초기 설정

```python
import matplotlib.pyplot as plt

plt.rcParams['font.family'] = 'D2Coding'
plt.rcParams['font.size'] = 15
plt.rcParams['figure.figsize'] = (10, 6)
plt.rcParams['axes.labelsize'] = 10
plt.rcParams['axes.grid'] = True
```

### 소계 그래프 표현

```python
pop_result['소계'].sort_values().plot(kind = 'barh')
plt.show()
```

### CCTV 비율 칼럼 추가 및 그래프 표현
#### CCTV 비율 칼럼 = 소계 / 인구수 * 100

```python
pop_result['CCTV 비율'] = (pop_result['소계'] / pop_result['인구수']) * 100
pop_result
```

#### CCTV 칼럼 막대 그래프 표현

```python
pop_result['CCTV 비율'].plot(kind = 'barh', figsize = (10, 10), color = ['red', 'orange', 'yellow', 'green', 'blue'])
plt.show()
```

#### 인구수별 카메라 점 그래프 표현

```python
plt.scatter(data_result['인구수'], data_result['소계'], s = 50)
plt.xlabel('인구수')
plt.ylabel('CCTV')
plt.title('인구수별 카메라')
plt.show()
```

#### 인구수별 카메라 점 그래프에 선 그래프 추가

```python
fp = np.polyfit(data_result['인구수'], data_result['소계'], 1)
fp
```

```python
fp1 = np.poly1d(fp)
fx = np.linspace(100000, 700000, 100)
```

```python
plt.scatter(data_result['인구수'], data_result['소계'], s = 50)
plt.plot(fx, fp1(fx), ls = 'dashed', lw = 3, color = 'green')
plt.xlabel('인구수')
plt.ylabel('CCTV')
plt.title('인구수별 카메라')
plt.show()
```
